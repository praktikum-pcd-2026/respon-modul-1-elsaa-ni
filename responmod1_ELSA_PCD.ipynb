```python
def fungsi_empat_R(image, val): 
    h, w = image.shape 
    val = int(val) % 360  

    if val == 0: 
        return image 

    elif val == 90: 
        rot = np.zeros((w, h), dtype=image.dtype) 
        for i in range(h): 
            for j in range(w): 
                rot[j, h - 1 - i] = image[i, j] 
        return rot 

    elif val == 180: 
        rot = np.zeros_like(image, dtype=image.dtype) 
        for i in range(h): 
            for j in range(w): 
                rot[h - 1 - i, w - 1 - j] = image[i, j] 
        return rot 

    elif val == 270:  
        rot = np.zeros((w, h), dtype=image.dtype) 
        for i in range(h): 
            for j in range(w): 
                rot[w - 1 - j, i] = image[i, j] 
        return rot 

    return image
```

4. Rotasi
Fungsi fungsi_empat_R digunakan untuk memutar citra pada sudut tertentu, yaitu 0°, 90°, 180°, atau 270°. Citra dipandang sebagai sebuah matriks, kemudian posisi setiap piksel diubah melalui proses perulangan sesuai dengan sudut rotasi yang dipilih. Pada rotasi 90° dan 270°, dimensi matriks berubah karena tinggi dan lebar saling ditukar, sedangkan pada rotasi 180° ukuran matriks tetap sama. Jika sudut yang diberikan tidak termasuk dalam pilihan tersebut, maka citra akan dikembalikan tanpa perubahan. Fungsi ini merupakan bentuk implementasi rotasi citra secara manual tanpa memanfaatkan fungsi bawaan dari OpenCV.

Kode tersebut digunakan untuk menerapkan proses rotasi pada citra, kemudian menampilkan hasilnya dengan bantuan Matplotlib. Citra buggy1 diputar sebesar 90° dan disimpan dalam variabel hasil1, lalu ditampilkan pada subplot pertama menggunakan plt.subplot(1,2,1). Selanjutnya, citra buggy2 diputar sebesar 270° dan disimpan dalam hasil2, kemudian ditampilkan pada subplot kedua.

Penggunaan parameter cmap='gray' bertujuan karena citra yang ditampilkan berupa grayscale, sedangkan plt.axis('off') digunakan untuk menghilangkan sumbu koordinat agar tampilan lebih bersih. Namun, karena plt.show() dipanggil setelah masing-masing subplot, kedua gambar akan ditampilkan secara terpisah, bukan dalam satu tampilan yang sama.

```python
hasil1 = fungsi_empat_R(buggy1, 90)
plt.subplot(1,2,1)
plt.imshow(hasil1, cmap='gray')
plt.axis('off')
plt.show()

hasil2 = fungsi_empat_R(buggy2, 270)
plt.subplot(1,2,2)
plt.imshow(hasil2, cmap='gray')
plt.axis('off')
plt.show()
```

5. Flip
Kode tersebut digunakan untuk melakukan pencerminan (flip) citra secara manual. Fungsi fungsi_satu memiliki dua opsi, yaitu 'X' untuk membalik citra secara horizontal (kiri–kanan) dan 'Y' untuk membalik citra secara vertikal (atas–bawah). Proses ini dilakukan dengan memanipulasi posisi piksel menggunakan array dari NumPy.

Hasilnya, citra buggy3 dibalik secara vertikal, sedangkan buggy4 dibalik secara horizontal. Kedua citra kemudian ditampilkan menggunakan Matplotlib dalam format grayscale tanpa menampilkan sumbu koordinat.

```python
def fungsi_satu(image, mode='X'): 
    h, w = image.shape 
    hasil = np.zeros((h, w), dtype=image.dtype) 
    for i in range(h): 
        for j in range(w): 
            if mode == 'X': 
                hasil[i, j] = image[i, w - 1 - j] 
            elif mode == 'Y': 
                hasil[i, j] = image[h - 1 - i, j] 
    return hasil
hasil3 = fungsi_satu(buggy3, 'Y')
hasil4 = fungsi_satu(buggy4, 'X')

plt.imshow(hasil3, cmap='gray')
plt.axis('off')
plt.show()

plt.imshow(hasil4, cmap='gray')
plt.axis('off')
plt.show()
```

6. Crop
Kode tersebut digunakan untuk melakukan proses cropping (pemotongan) citra secara otomatis dengan cara menghilangkan bagian kosong atau area berwarna hitam pada gambar. Fungsi fungsi_crop bekerja dengan mendeteksi baris dan kolom yang memiliki nilai piksel lebih dari nol menggunakan NumPy, kemudian menentukan batas atas, bawah, kiri, dan kanan dari area yang mengandung objek.

Setelah batas-batas tersebut diperoleh, citra dipotong sesuai area aktif sehingga hanya bagian yang memuat informasi yang ditampilkan. Proses cropping ini kemudian diterapkan pada beberapa citra hasil sebelumnya (hasil1 hingga hasil4) dan ditampilkan menggunakan Matplotlib dalam mode grayscale (cmap='gray'). Perintah untuk menyembunyikan sumbu juga digunakan agar tampilan lebih rapi dan fokus pada objek utama.

```python
def fungsi_crop(image):
    h, w = image.shape
    rows = np.any(image > 0, axis=1)
    cols = np.any(image > 0, axis=0)
    rmin, rmax = np.where(rows)[0][[0, -1]]
    cmin, cmax = np.where(cols)[0][[0, -1]]
    return image[rmin:rmax+1, cmin:cmax+1]

bgy1 = fungsi_crop(hasil1)
plt.imshow(bgy1, cmap='gray')
plt.axis('off')
plt.show()

bgy2 = fungsi_crop(hasil2)
plt.imshow(bgy2, cmap='gray')
plt.axis('off')
plt.show()

bgy3 = fungsi_crop(hasil3)
plt.imshow(bgy3, cmap='gray')
plt.axis('off')
plt.show()

bgy4 = fungsi_crop(hasil4)
plt.imshow(bgy4, cmap='gray')
plt.axis('off')
plt.show()
```

Kode tersebut digunakan untuk menampilkan citra hasil cropping dalam satu tampilan dengan memanfaatkan fitur subplot pada Matplotlib. Selanjutnya, fungsi plt.subplot(2,2,n) digunakan untuk membagi kanvas menjadi empat bagian (2 baris dan 2 kolom), sehingga setiap citra hasil cropping (bgy1, bgy3, bgy2, dan bgy4) dapat ditempatkan pada posisi yang berbeda.

Masing-masing citra ditampilkan dalam format grayscale dengan parameter cmap='gray', sementara plt.axis('off') digunakan untuk menghilangkan sumbu koordinat agar tampilannya lebih bersih. Terakhir, plt.show() dipanggil untuk menampilkan seluruh subplot secara bersamaan dalam satu jendela.

```python
def gabung_samping(citra_1, citra_2):
    c1, c2 = np.array(citra_1), np.array(citra_2)
    
    tinggi = max(c1.shape[0], c2.shape[0])
    lebar = c1.shape[1] + c2.shape[1]
    
    hasil = np.zeros((tinggi, lebar), dtype=c1.dtype)
    
    hasil[0:c1.shape[0], 0:c1.shape[1]] = c1
    hasil[0:c2.shape[0], c1.shape[1]:] = c2
    
    return hasil

def gabung_bawah(citra_1, citra_2):
    c1, c2 = np.array(citra_1), np.array(citra_2)
    
    tinggi = c1.shape[0] + c2.shape[0]
    lebar = max(c1.shape[1], c2.shape[1])
    
    hasil = np.zeros((tinggi, lebar), dtype=c1.dtype)
    
    hasil[0:c1.shape[0], 0:c1.shape[1]] = c1
    hasil[c1.shape[0]:, 0:c2.shape[1]] = c2
    
    return hasil
```

Kode tersebut digunakan untuk menggabungkan dua citra menjadi satu dalam dua arah, yaitu secara horizontal dan vertikal. Fungsi gabung_samping berfungsi untuk menggabungkan citra ke arah samping, dengan menempatkan citra pertama di sisi kiri dan citra kedua di sisi kanan. Tinggi citra hasil mengikuti citra yang paling tinggi, sedangkan lebarnya merupakan hasil penjumlahan dari kedua lebar citra.

Sementara itu, fungsi gabung_bawah digunakan untuk menggabungkan citra secara vertikal, di mana citra pertama diletakkan di bagian atas dan citra kedua di bagian bawah. Tinggi citra hasil merupakan gabungan dari tinggi kedua citra, sedangkan lebarnya menyesuaikan dengan citra yang paling lebar. Kedua fungsi ini memanfaatkan array dari NumPy untuk membuat matriks baru dan menempatkan masing-masing citra pada posisi yang sesuai.

Kode tersebut berfungsi untuk menghilangkan bagian tepi citra, lalu menggabungkan beberapa citra menjadi satu tampilan utuh. Fungsi hapus_pinggir bekerja dengan memotong sisi-sisi citra sebanyak n piksel dari setiap arah menggunakan teknik slicing array pada NumPy. Proses ini diterapkan pada citra bgy1 hingga bgy4 untuk menghapus bagian tepi yang tidak diperlukan.

Selanjutnya, citra-citra tersebut digabungkan secara bertahap. Citra bgy1 dan bgy3 digabung secara horizontal membentuk bagian atas, sementara bgy2 dan bgy4 digabung secara horizontal menjadi bagian bawah. Kedua hasil tersebut kemudian digabung secara vertikal sehingga menghasilkan satu citra utuh yang dinamakan gabungan.

Hasil akhir ditampilkan menggunakan Matplotlib dalam mode grayscale, dengan sumbu koordinat yang disembunyikan agar tampilan terlihat lebih bersih dan fokus pada citra.

```python
def hapus_pinggir(img, n=1):
    return img[n:-n, n:-n]

bgy1 = hapus_pinggir(bgy1, 1)
bgy2 = hapus_pinggir(bgy2, 1)
bgy3 = hapus_pinggir(bgy3, 1)
bgy4 = hapus_pinggir(bgy4, 1)

atas = gabung_samping(bgy1, bgy3)

bawah = gabung_samping(bgy2, bgy4)

gabungan = gabung_bawah(atas, bawah)

plt.imshow(gabungan, cmap='gray')
plt.axis('off')
plt.show()
```

Kode tersebut digunakan untuk melakukan operasi pembagian nilai intensitas piksel pada citra sehingga menghasilkan tampilan yang lebih gelap. Fungsi pembagian2 bekerja dengan membagi setiap nilai piksel pada citra dengan suatu konstanta (default = 2) menggunakan perulangan serta array dari NumPy. Hasil dari proses ini disimpan dalam variabel hasil dengan tipe data uint8.

Selanjutnya, citra gabungan diproses menggunakan fungsi tersebut dan hasilnya disimpan dalam variabel dibagi2. Citra hasil kemudian ditampilkan menggunakan Matplotlib dalam format grayscale.

```python
def pembagian2(citra, dibagi=2):
    tinggi, lebar = citra.shape
    hasil = np.zeros((tinggi, lebar), dtype=np.uint8)
    for i in range(tinggi):
        for j in range(lebar):
            hasil[i, j] = int(citra[i, j] / dibagi)
    return hasil

dibagi2 = pembagian2(gabungan)

plt.imshow(dibagi2, cmap='gray')

plt.show()
```

Kode tersebut digunakan untuk menghitung sekaligus menampilkan histogram dari citra grayscale. Fungsi hist bekerja dengan membuat array berukuran 256, yang mewakili rentang nilai intensitas piksel dari 0 hingga 255. Selanjutnya, jumlah kemunculan setiap nilai piksel dihitung melalui perulangan pada seluruh elemen citra menggunakan NumPy, dan hasilnya disimpan dalam variabel gambarhisto.

Histogram kemudian divisualisasikan dalam bentuk diagram batang menggunakan Matplotlib, di mana sumbu X merepresentasikan nilai intensitas piksel, sedangkan sumbu Y menunjukkan jumlah kemunculan piksel tersebut. Visualisasi ini berguna untuk memahami distribusi tingkat kecerahan pada citra.

```python
def hist(citra):
    histogram = np.zeros(256)
    for i in range(citra.shape[0]):
        for j in range(citra.shape[1]):
            histogram[int(citra[i, j])] += 1
    return histogram


gambarhisto = hist(dibagi2)

plt.bar(range(256), gambarhisto, color='blue')
plt.title("Histogram")
plt.xlabel("Nilai Intensitas")
plt.ylabel("Jumlah Piksel")
plt.show()
```

```python
def normalisasi(image):
    min_val = np.min(image)
    max_val = np.max(image)

    if max_val - min_val == 0:
        return image

    norm = (image - min_val) / (max_val - min_val) * 255
    return norm.astype(np.uint8)
```

Fungsi normalisasi digunakan untuk menyesuaikan intensitas citra agar nilai piksel berada pada rentang 0–255. Langkah awalnya adalah mencari nilai minimum (min_val) dan maksimum (max_val) dari citra menggunakan NumPy. Jika selisih antara keduanya bernilai nol, maka citra dikembalikan apa adanya untuk menghindari pembagian dengan nol.

Jika tidak, setiap piksel akan diskalakan menggunakan rumus normalisasi sehingga nilainya tersebar secara merata dalam rentang 0–255. Hasil akhir kemudian dikonversi ke tipe data uint8 agar sesuai dengan format citra grayscale.

Kode tersebut digunakan untuk menerapkan proses normalisasi pada citra sekaligus menampilkan histogram hasilnya. Citra dibagi2 dinormalisasi menggunakan fungsi normalisasi sehingga nilai pikselnya kembali tersebar dalam rentang 0–255, lalu disimpan dalam variabel normal.

Selanjutnya, histogram dari citra yang telah dinormalisasi dihitung menggunakan fungsi hist dan disimpan dalam variabel gambarhisto_norm. Histogram tersebut kemudian ditampilkan dalam bentuk diagram batang menggunakan Matplotlib, dengan warna hijau untuk menggambarkan distribusi intensitas piksel setelah proses normalisasi. Visualisasi ini membantu dalam menganalisis peningkatan kontras pada citra.

```python
normal = normalisasi(dibagi2)

gambarhisto_norm = hist(normal)

plt.bar(range(256), gambarhisto_norm, color='green')
plt.title("Histogram Setelah Normalisasi")
plt.xlabel("Nilai Intensitas")
plt.ylabel("Jumlah Piksel")
plt.show()
```

```python
def zero_padding(citra, pad):
    h, w = citra.shape
    hasil = np.zeros((h + 2*pad, w + 2*pad))
    hasil[pad:pad+h, pad:pad+w] = citra
    return hasil

kernel = np.array([
    [-1/10, -1/10, -1/10, -1/10],
    [-1/10,  6/10,  6/10, -1/10],
    [-1/10,  6/10,  6/10, -1/10],
    [-1/10, -1/10, -1/10, -1/10]
])

def konvolusi(citra, kernel):
    kh, kw = kernel.shape
    pad = kh // 2
    citra_pad = zero_padding(citra, pad)

    h, w = citra.shape
    hasil = np.zeros((h, w))

    for i in range(h):
        for j in range(w):
            area = citra_pad[i:i+kh, j:j+kw]
            hasil[i, j] = np.sum(area * kernel)

    return hasil
```

Kode tersebut digunakan untuk melakukan proses konvolusi pada citra menggunakan kernel tertentu. Fungsi zero_padding bertugas menambahkan padding berupa nilai nol di sekeliling citra, sehingga ukuran output tetap sama setelah proses konvolusi. Padding ini dibuat dengan memanfaatkan array dari NumPy untuk menambahkan bingkai nol di sekitar citra.

Kernel yang digunakan berukuran 4×4 dengan kombinasi nilai negatif dan positif, yang bertujuan untuk menonjolkan fitur tertentu pada citra, seperti penajaman atau deteksi bagian tertentu. Selanjutnya, fungsi konvolusi bekerja dengan menggeser kernel ke seluruh area citra. Pada setiap posisi, dilakukan perkalian antara nilai piksel pada bagian citra dengan nilai kernel, kemudian hasilnya dijumlahkan untuk menghasilkan satu nilai piksel baru.

Proses ini diulangi hingga mencakup seluruh bagian citra, sehingga menghasilkan citra baru sebagai hasil dari proses konvolusi.

Kode tersebut digunakan untuk menerapkan proses konvolusi pada citra sekaligus membandingkan hasil sebelum dan sesudah pemrosesan. Citra normal terlebih dahulu diproses menggunakan fungsi konvolusi dengan kernel yang telah ditentukan. Hasilnya kemudian dinormalisasi kembali menggunakan fungsi normalisasi agar nilai piksel berada dalam rentang 0–255, dan disimpan dalam variabel citra_konv_norm.

Selanjutnya, kedua citra ditampilkan menggunakan Matplotlib dalam satu tampilan yang terdiri dari dua subplot. Subplot pertama menunjukkan citra sebelum proses konvolusi, sedangkan subplot kedua menampilkan citra setelah konvolusi. Penggunaan axis on bertujuan untuk menampilkan sumbu koordinat sehingga perbedaan antara kedua citra dapat diamati dengan lebih jelas.

```python
citra_konv = konvolusi(normal, kernel)

citra_konv_norm = normalisasi(citra_konv)

plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.imshow(normal, cmap='gray')
plt.title("Sebelum Konvolusi")
plt.axis('on')

plt.subplot(1,2,2)
plt.imshow(citra_konv_norm, cmap='gray')
plt.title("Setelah Konvolusi")
plt.axis('on')

plt.show()
```
